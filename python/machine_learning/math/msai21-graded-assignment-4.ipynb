# Midpoint rule

```python
import math
```

```python
def midpoint(func, a, b, n):
    h = (b - a) / n
    res = 0
    for i in range(n):
        res += func((a + h/2) + i*h) * h
    return res
```

```python
func = lambda x: math.sin(x**2)
midpoint(func, -2, 4, 6)
```

# Trapezoidal rule

```python
def trapezoid(func, a, b, n):
    h = (b - a) / n
    res = 0
    for i in range(n):
        res += (func(a + i*h) + func(a + (i+1)*h))/2 * h
    return res
```

```python
func = lambda x: math.sin(x**2)
trapezoid(func, -2, 4, 6)
```
