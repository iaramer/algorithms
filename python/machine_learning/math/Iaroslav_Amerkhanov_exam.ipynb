### Definitions:

**Convex function** - a function $f$: $\mathbb{R}^n → \mathbb{R}$ is convex if its domain $D$ is a convex set $∀ a_1, a_2 \in D : \forall \lambda \in [0,1]:$

$$f( \lambda a_1 + (1-\lambda) a_2) \leq \lambda f(a_1) +(1-\lambda)f(a_2)$$

**Strongly convex function** - a function $f$ is called strongly convex with $m>0$ and domain $D$ if $\forall a_1,a_2 \in D$:

$$(\nabla f(a_1)-\nabla f(a_2))^\top (a_1-a_2) \geq m||a_1-a_2||_2^2$$

### Example:

**Strongly convex function:**

$$f(x)=||x||_2^2$$

**The constant of strong convexity:**

$$\nabla f(x)=2x$$

$$2(x_1-x_2)^\top (x_1-x_2) \geq m||x_1-x_2||_2^2 ⇒ m=2$$

**The definition of the convex set:**

The set C is called convex if $\forall a_1,a_2 \in C,\forall θ \in [0,1]:$ 

$$\theta a_1 +(1-\theta)a_2 \in C$$

**Check if the following set  G  is convex:**

Let's say that $a,b \in G$ and $θ \in [0,1]$

Thus, if the set is convex, the following statement must be true:

$$\theta a +(1-\theta)b \in G\ \forall\ θ \in [0,1]$$

In other words:

$$
A_0 + (\theta a_1+(1-\theta)b_1)A_1 + \ldots + (\theta a_n+(1-\theta)b_n)A_n \succeq 0
$$

Let's rearrange the elements:

$$
A_0 + (\theta a_1+(1-\theta)b_1)A_1 + \ldots + (\theta a_n+(1-\theta)b_n)A_n \succeq 0
$$

As A_0 is PSD (by condition):
$$
A_0 = \theta A_0 + (1-\theta) A_0
$$

Thus,

$$
\theta A_0 + (1-\theta) A_0 + \theta a_1 A_1+(1-\theta)b_1 A_1 + \ldots + \theta a_n A_n+(1-\theta)b_n A_n \succeq 0
$$

$$
\theta (A_0 + a_1 A_1 + \ldots + a_n A_n) + (1-\theta) (A_0 + b_1 A_1 + \ldots + b_n A_n) \succeq 0
$$

$$
\theta (\succeq 0) + (1-\theta) (\succeq 0) \succeq 0
$$

Which is true.

Finally, we can conclude that the given set is convex.

**Is this problem smooth or not? Explain your answer**

The smoothness of function is defined as the function is differentiable. As the $max$ function isn't differentiable, we can conclude that the given function isn't smooth.

**Is this problem convex or not? Explain your answer**

An optimization problem is convex if both the objective function and its domain are convex. In the given example we are given only the objective function:

$$
f(w) = \frac{1}{2}\|w\|_2^2 + \sum_{i=1}^m \max(0, 1 - y_iw^{\top}x_i)

So, we have the following:

1. Square of a norm is a convex function
2. Max function is also convex

4. Sum is a convex function
5. Finally, the weighted sum of two elements is also convex

Thus, the problem is convex.

**If this problem is not smooth, transform it into equivalent smooth problem**

We can do this by turning our unconstraint optimization problem into the constraint one:

$$
\min_w \frac{1}{2}\|w\|_2^2 + m - \sum_{i=1}^m y_iw^{\top}x_i
\\
s.t.\ y_iw^{\top}x_i-1 \leq 0\ \forall i \in \{1, \ldots, m\}

**Write the KKT conditions**

Let's state the Lagrangian:

$$
L = \frac{1}{2}\|w\|_2^2 + m - \sum_{i=1}^m y_iw^{\top}x_i+ \sum_{i=1}^m \mu_i (y_iw^{\top}x_i-1)

$$
\frac{\partial L}{\partial w} = w - \sum_{i=1}^m y_i x_i + \sum_{i=1}^m \mu_i y_i x_i

$$
1)\ y_iw^{\top}x_i-1 \leq 0
\\
2)\ \mu_i \geq 0

3)\ \mu_i(y_iw^{\top}x_i-1) =0

4)\  w - \sum_{i=1}^m y_i x_i + \sum_{i=1}^m \mu_i y_i x_i = 0
$$

The necessary conditions are also sufficient for optimality, if the objective function is a convex function. Thus, the KKT conditions are sufficient for the given minimization problem.

Let's derive the necessary formulas:

$$
\nabla \log (e^x + e^{-x}) = \frac{e^x-e^{-x}}{e^x+e^{-x}} = \frac{e^{2x}-1}{e^{2x}+1}
$$

$$
H = \frac{2e^{2x} (1+ e^{2x}) - 2e^{2x}(e^{2x}-1)}{(e^{2x}+1)^2} = \frac{4e^{2x}}{(e^{2x}+1)^2}
$$

```python
# Your answer is here
import numpy as np
import math
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Implement and run Newton method with constant step size  α=1  and initial 
# guesses  x0=1  and  x0=1.1

def func(x):
    return np.log(np.exp(x) + np.exp(-x)) 

def grad(x):
    return (np.exp(2*x)-1) / (np.exp(2*x)+1)

def hess(x):
    return 4 * np.exp(2*x) / ((np.exp(2*x)+1)**2)

def newton(x0, alpha=1, tol=1e-9, n_iter=1000, verbose=False):
    x = x0
    history = []
    for i in range(1, n_iter+1):
        if verbose:
            print(f"Gradient: {grad(x)}; Hessian: {hess(x)}")
        hess_v = hess(x)
        if hess_v == 0:
            print(f'Devision by zero (H=0) at step {i}. Check for divergence.')
            break
        h = - grad(x) / hess_v
        if math.isnan(h):
            print(f'NaN obtained at step {i}. Check for divergence.')
            break
        x += alpha * h
        err = np.linalg.norm(grad(x))
        history.append((x, err))
        if err < tol:
            print(f'Converged successfully in {i} steps.')
            break
    return history
```

First of all, let's show, what we're trying to achieve. We're trying to find the point, where the gradient is equal to zero:

$$\nabla f(x) = \nabla(\frac{1}{2}x^{\top}Ax - b^{\top}x)
 = \frac{1}{2}((A + A^\top) x - b) = 0$$

For simplicity we'll imply that A is also a symmetric matrix.

$$Ax-b=0$$

$$Ax=b$$

Let's find the step size.

$$\nabla f(x_{k+1})^\top \nabla f(x_{k}) = 0$$

$$(Ax_{k+1}-b)^\top (Ax_{k}-b) = 0$$

$$(A(x_{k}-\eta (Ax_k-b))-b)^\top (Ax_{k}-b) = 0$$

$$(Ax_{k}-\eta A(Ax_k-b)-b)^\top (Ax_{k}-b) = 0$$

$$(Ax_{k}-\eta AAx_k-\eta Ab-b)^\top (Ax_{k}-b) = 0$$

$$(Ax_k)^\top Ax_k - \eta (AAx_k)^\top Ax_k + \eta (Ab)^\top Ax_k - b^\top Ax_k-(Ax_k)^\top b + \eta (AAx_k)^\top b -\eta(Ab)^\top b +b^\top b = 0$$

$$\eta ((AAx_k-Ab)Ax_k-(AAx_k-Ab)^\top b) = (Ax_k-b)^\top (Ax_k-b)$$

$$\eta = \frac{(Ax_k-b)^\top (Ax_k-b)}{(A(Ax_k-b))^\top (Ax_k-b)}$$

$$\eta = \frac{\nabla f(x_k)^\top \nabla f(x_k)}{\nabla f(x_k)^\top A \nabla f(x_k)}$$

```python
# Run gradient descent with three step size selection trategies: steepest 
# descent rule, optimal constant step size and Armijo rule.

import numpy as np
import matplotlib.pyplot as plt
import liboptpy.unconstr_solvers as methods
from liboptpy.unconstr_solvers.fo import GradientDescent, AcceleratedGD
from liboptpy.step_size import Backtracking, ConstantStepSize
from liboptpy.base_optimizer import LineSearchOptimizer


def func(x, A, b):
    return 0.5 * np.linalg.norm(A @ x - b)**2

def grad(x, A, b):
    return A.T @ (A @ x - b)
```

Let's describe the step size we're searching for in the Steepest GD.

$$x_{k+1}=x_k+h p_k$$

Where $h$ is the step size, k is the number of iteration and p_k is gradient on the k-th step.

$$\frac{1}{2}||Ax-b||_2^2 = \frac{1}{2}||A(x_k+h p_k)-b||_2^2 $$

$$\frac{\partial }{\partial h} \frac{1}{2}||A(x_k+h p_k)-b||_2^2 = 0$$

$$2 A^\top (Ax_k+Ahp_k-b)p_k=0$$

$$h = -\frac{A^\top p_k (Ax_k-b)}{p_k^\top A^\top A p_k} = -\frac{p_k^\top p_k}{p_k^\top A^\top A p_k}$$

```python
%time
# Steepest descent rule
A = np.random.randn(1000, 100)
b = np.random.randn(1000)
x = np.random.randn(100)

def gd_steepest(x0, A, b, tol=1e-4, n_iter=1000):
    x = x0
    res = []
    for i in range(1, n_iter + 1):
        g = grad(x, A, b)
        alpha = (g.T @ g) / (g.T @ A.T @ A @ g)
        x -= alpha * g
        res.append((x, np.linalg.norm(g)))
        if np.linalg.norm(grad(x, A, b)) < tol:
            print(f"Method converged at step {i}")
            break
    if i == n_iter:
        print(f"Maximum bumber of iterations ({n_iter}) exceeded.")
    return res

res_stgd = gd_steepest(x, A, b)
conv_sd = [float(k[1]) for k in res_stgd]
print(f"Minimal value: {func(res_stgd[-1][0], A, b)}")
```

```python
%time
# Optimal constant step size
A = np.random.randn(1000, 100)
b = np.random.randn(1000)
x = np.random.randn(100)

optimal_step = 1 / max(np.linalg.eigvals(A.T @ A))

gd = GradientDescent(lambda x_: func(x_, A, b), 
                     lambda x_: grad(x_, A, b), 
                     ConstantStepSize(optimal_step))
gd.solve(x, tol=1e-4, max_iter=1000, disp=True)

conv_os = [np.linalg.norm(grad(x_, A, b)) for x_ in gd.get_convergence()]
```

```python
%time
# Armijo rule
A = np.random.randn(1000, 100)
b = np.random.randn(1000)
x = np.random.randn(100)

gd = GradientDescent(lambda x_: func(x_, A, b), 
                     lambda x_: grad(x_, A, b), 
                     Backtracking("Armijo", rho=0.5, beta=0.25, init_alpha=10))
gd.solve(x, tol=1e-4, max_iter=1000, disp=True)

conv_arm = [np.linalg.norm(grad(x_, A, b)) for x_ in gd.get_convergence()]
```

```python
# Make convergence plots
plt.figure(figsize=(10,6))
plt.xlabel("Step", fontsize=20)
plt.ylabel("Error", fontsize=20)
plt.title("Convergence plot", fontsize=20)
plt.semilogy(conv_sd, label="Stepest Descent")
plt.semilogy(conv_os, label="Optimal constant step size")
plt.semilogy(conv_arm, label="Armijo rule")
plt.legend()
plt.grid(True)
plt.show()
```

```python
%time
# Heavy ball method
A = np.random.randn(1000, 100)
b = np.random.randn(1000)
x = np.random.randn(100)


A_eig = np.linalg.eigvals(A.T @ A)

mu = np.min(A_eig)
L =  np.max(A_eig)
alpha_opt = 4 / (np.sqrt(L) + np.sqrt(mu))**2
beta_opt = np.maximum((1 - np.sqrt(alpha_opt * L))**2, 
                      (1 - np.sqrt(alpha_opt * mu))**2)

class HeavyBall(LineSearchOptimizer):
    def __init__(self, f, grad, step_size, beta, **kwargs):
        super().__init__(f, grad, step_size, **kwargs)
        self._beta = beta
    
    def get_direction(self, x):
        self._current_grad = self._grad(x)
        return -self._current_grad

    def _f_update_x_next(self, x, alpha, h):
        if len(self.convergence) < 2:
            return x + alpha * h
        else:
            return x + alpha * h + self._beta * (x - self.convergence[-2])
    
    def get_stepsize(self):
        return self._step_size.get_stepsize(self._grad_mem[-1], 
                                            self.convergence[-1], 
                                            len(self.convergence))

gd = HeavyBall(lambda x_: func(x_, A, b), lambda x_: grad(x_, A, b), 
               ConstantStepSize(alpha_opt), beta=beta_opt)
x = gd.solve(x, tol=1e-4, max_iter=1000, disp=True)

x_conv = gd.get_convergence()
hb_conv = [np.linalg.norm(grad(x, A, b)) for x in x_conv]
```
