```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/throwsh.csv")
```

```python
df
```

```python
def first(df):
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df['date'].dt.year
    dfg = df.groupby(['year'])
    res_df = df.set_index('vol').loc[list(dfg['vol'].max())].reset_index()[['date', 'vol']]
    # remove duplicates:
    res_df['year'] = res_df['date'].dt.year
    res_df = res_df.drop_duplicates(subset=['year']).drop('year', axis=1)
    return res_df
```

```python
def second(df):
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df['date'].dt.year
    dfg = df.groupby(['year'])
    dfg = pd.DataFrame(dfg['close'].max()).reset_index()
    res_df = pd.merge(dfg, df, on=['year','close'])[['date','close']]
    return res_df
```

```python
# def solution(files: List[str]) -> List[List[pd.DataFrame, pd.DataFrame]]:
def solution(files):
    # files - any of available files, i.e:
    # files = ["./data/framp.csv", "./data/gnyned.csv", "./data/gwoomed.csv",
    #            "./data/hoilled.csv", "./data/plent.csv", "./data/throwsh.csv",
    #            "./data/twerche.csv", "./data/veeme.csv"]
    res = []
    for file in files:
        df = pd.read_csv(file)
        res.append([def first(df), def second(df)])
    return res
```
