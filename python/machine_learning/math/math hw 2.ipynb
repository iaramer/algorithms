# Problem 2

Find the straight line y = w_1*x + w_0 that is the best least-squares fit to the points (0,0), (1,3) and (2,7).

### Solution

The general formula for the leasr squares solution is the following:

(X.T * X) ^ (-1) * X.T * Y

```python
X = np.array([
              [0, 1, 2],
              [1, 1, 1]
              ]).T
y = np.array([0, 3, 7])
```

```python
X
```

```python
y
```

```python
inv = np.linalg.inv
```

```python
w = inv(X.T @ X) @ X.T @ y
w
```

Thus, the straight line is **y = 3.5x - 1/6**

### Let's check it!

```python
w1, w0 = np.linalg.lstsq(X, y, rcond=None)[0]
w1, w0
```
